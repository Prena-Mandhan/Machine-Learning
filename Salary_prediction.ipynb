```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error,root_mean_squared_error
```

Step 1 Import libraries and load the data

```python
df = pd.read_csv("Salary_Data.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

STEP 2 #SPLIT THE DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(df['YearsExperience'],df['Salary'],test_size=0.2, random_state=42)
```

```python
print(X_train)
print(Y_train)
```

```python
print(X_train.shape)
print(Y_train.shape)

print(X_test.shape)
print(Y_test.shape)
```

```python
print("Values of X_train = ",X_train.values)
print("Values of Y_train = ",Y_train.values)

print("Values of X_test = ",X_test.values)
print("Values of X_test = ",Y_test.values)
```

```python
x_train = X_train.values.reshape(-1,1)
y_train = Y_train.values.reshape(-1,1)
x_test= X_test.values.reshape(-1,1)
y_test= Y_test.values.reshape(-1,1)
```

```python
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

STEP 3 #TRAIN THE MODEL

```python
lr= LinearRegression()
lr.fit(x_train,y_train)
```

STEP 4 AND STEP 5 #PREDICT AND EVALUATE

```python
y_pred = lr.predict(x_test)
x_test
```

```python
print("Prediction of this model ",r2_score(y_test,y_pred))
print("average absolute difference between actual and predicted values ",mean_absolute_error(y_test, y_pred))
```

```python
m = lr.coef_
c= lr.intercept_

x= 1.1	
y = m*x+c
y
```
